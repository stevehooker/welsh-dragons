```python
#@title Mount Google Drive
#@markdown Mount your Google Drive to access the ComfyUI installation

print("Mounting Google Drive...")
from google.colab import drive
drive.mount('/content/drive')
print("Google Drive mounted!")
```

```python
#@title Download Checkpoints
#@markdown Select which models to download

DOWNLOAD_FLUX2_dev = True  #@param {type:"boolean"}
DOWNLOAD_SDXL = True  #@param {type:"boolean"}
DOWNLOAD_SD15 = False  #@param {type:"boolean"}
DOWNLOAD_QWEN_EDIT = True  #@param {type:"boolean"}

print("=== Downloading Checkpoints ===")

if DOWNLOAD_FLUX2_dev:
    print("Downloading FLUX.2-dev (~6GB)...")
    print("Note: Requires HuggingFace authentication")
    
    try:
        from google.colab import userdata
        HF_TOKEN = userdata.get('HF_TOKEN')
        
        # Download with authentication
        !wget -c --header="Authorization: Bearer {HF_TOKEN}" \
            "https://huggingface.co/black-forest-labs/FLUX.2-dev/resolve/main/flux2-dev.safetensors" \
            -O /content/drive/MyDrive/ComfyUI/models/checkpoints/flux2-dev.safetensors
        print("FLUX.2-dev downloaded!")
    except Exception as e:
        print("\n‚ö†Ô∏è FLUX.2-dev requires authentication!")
        print("\nTo download:")
        print("1. Get a HuggingFace token: https://huggingface.co/settings/tokens")
        print("2. Accept FLUX.2 terms: https://huggingface.co/black-forest-labs/FLUX.2-dev")
        print("3. In Colab: Click the key icon (üîë) on left sidebar")
        print("4. Add a secret named 'HF_TOKEN' with your token")
        print("5. Enable notebook access for this secret")
        print("6. Re-run this cell")
        print(f"\nError: {e}")
if DOWNLOAD_SDXL:
    print("Downloading SDXL base (~6.9GB)...")
    !wget -c https://huggingface.co/stabilityai/stable-diffusion-xl-base-1.0/resolve/main/sd_xl_base_1.0.safetensors -P /content/drive/MyDrive/ComfyUI/models/checkpoints/
    print("SDXL base downloaded!")

if DOWNLOAD_SD15:
    print("Downloading SD 1.5 (~4GB)...")
    !wget -c https://huggingface.co/runwayml/stable-diffusion-v1-5/resolve/main/v1-5-pruned-emaonly.safetensors -P /content/drive/MyDrive/ComfyUI/models/checkpoints/
    print("SD 1.5 downloaded!")

if DOWNLOAD_QWEN_EDIT:
    print("Downloading Qwen-Image-Edit-2511 (split model, ~10GB total)...")
    # Install huggingface_hub if not already installed
    !pip install -q huggingface_hub
    
    # Download the entire transformer folder
    !huggingface-cli download Qwen/Qwen-Image-Edit-2511 \
        --include "transformer/*" \
        --local-dir /content/drive/MyDrive/ComfyUI/models/qwen-image-edit
    print("Qwen model downloaded to models/qwen-image-edit/transformer/")

print("")
print("=" * 60)
print("Downloads complete!")
print("=" * 60)
print("")
print("Models saved to:")
print("  ‚Ä¢ FLUX/SDXL/SD1.5: /content/drive/MyDrive/ComfyUI/models/checkpoints/")
print("  ‚Ä¢ Qwen Edit: /content/drive/MyDrive/ComfyUI/models/qwen-image-edit/")
```

## Model Information

### FLUX.2-dev (~6GB)
Latest FLUX model from Black Forest Labs. High-quality text-to-image generation.

**Requires HuggingFace authentication:**
1. Create a HuggingFace account
2. Get a token from [huggingface.co/settings/tokens](https://huggingface.co/settings/tokens)
3. Accept terms at [black-forest-labs/FLUX.2-dev](https://huggingface.co/black-forest-labs/FLUX.2-dev)
4. Add token as Colab secret named `HF_TOKEN`

### SDXL Base (~6.9GB)
Stable Diffusion XL base model. Industry standard for high-resolution image generation.

### SD 1.5 (~4GB)
Stable Diffusion 1.5. Fast and widely compatible, huge ecosystem of fine-tunes and LoRAs.

### Qwen-Image-Edit-2511 (~10GB)
Image editing model from Alibaba. Split across multiple files (automatically handled).

## Notes

- All models are downloaded to your Google Drive for persistence
- Downloads will resume if interrupted
- Split models (like Qwen) work automatically - no merging needed
- You can run this notebook independently to add more models later

## Setting up HuggingFace Token (for FLUX.2)

1. Click the **key icon** (üîë) in the left sidebar of Colab
2. Click **+ Add new secret**
3. Name: `HF_TOKEN`
4. Value: Your HuggingFace token
5. Toggle **Notebook access** ON for this notebook
6. The token is stored securely and never visible in the notebook
